```python
df.rename(columns={
    'Annual Income (k$)': 'Annual_Income',
    'Spending Score (1-100)': 'Spending_Score'
}, inplace=True)
```

```python
df.head()
```

#Gender Distribution

```python
sns.countplot(x = 'Gender',data=df)
plt.title(" Gender Distribution of Customers")
plt.show()
```

### Gender Distribution Insight
The dataset shows a fairly balanced distribution of male and female customers.
This indicates that the mall attracts customers from both genders almost equally,
making the analysis gender-neutral without strong bias toward a single group.


#Age Distribution

```python
sns.histplot(df['Age'], bins=20, kde=True)
plt.title('Age Distribution of Customers')
plt.show()
```

### Age Distribution Insight
The age distribution shows that most customers fall between the age group of
approximately 20 to 45 years. This suggests that young adults and middle-aged
customers are the primary visitors to the mall.
Fewer customers are observed in the higher age groups.


#Annual Income Distribution

```python
sns.histplot(df['Annual_Income'], bins=20, kde=True)
plt.title("Annual Income Distribution")
plt.show()
```

### Annual Income Distribution Insight
The annual income distribution indicates that most customers earn a moderate income,
with fewer customers at extremely low or very high income levels.
This suggests the mall primarily targets middle-income groups.


#Spending Score Distribution

```python
sns.histplot(df['Spending_Score'], bins=20, kde=True)
plt.title("Spending Score Distribution")
plt.show()
```

### Spending Score Distribution Insight
The spending score distribution is fairly spread across the range of values,
indicating varied purchasing behavior among customers.
Some customers spend conservatively, while others exhibit high spending tendencies.


#Income vs Spending Score

```python
plt.figure(figsize=(7,5))
sns.scatterplot(x='Annual_Income', y='Spending_Score', hue='Gender' , data=df)
plt.title('Annual Income vs Spending Score')
plt.show()
```

### Annual Income vs Spending Score Insight
The plot shows that high spending scores are not always associated with high income.
Some customers with moderate income also exhibit high spending behavior,
suggesting that income alone does not determine spending patterns.


#Age vs Spending Score

```python
plt.figure(figsize=(7,5))
sns.scatterplot(x= 'Age', y='Spending_Score', hue='Gender', data=df)
plt.title('Age vs Spending Score')
plt.show()
```

### Age vs Spending Score Insight
The scatter plot shows that younger customers tend to have higher spending scores,
while spending generally decreases as age increases.
This indicates that younger customers are more likely to spend actively compared
to older customers.
